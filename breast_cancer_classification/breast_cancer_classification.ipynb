```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
Breast_cancer = load_breast_cancer()
```

```python
Breast_cancer["data"] 
```

```python
Breast_cancer_df = pd.DataFrame(np.c_[Breast_cancer["data"], Breast_cancer["target"]], 
                                columns = np.append(Breast_cancer["feature_names"],["target"]))

Breast_cancer_df.head(10)
```

```python
Breast_cancer_df.shape
```

```python
Breast_cancer_df.columns
```

```python
sns.pairplot(Breast_cancer_df, hue="target", palette="gnuplot2" ,
             vars = ['mean radius', 'mean texture', 'mean perimeter','mean area','mean smoothness'])
```

```python
sns.countplot(x=Breast_cancer_df['target'],hue=Breast_cancer_df['target'], palette="gnuplot2")
```

```python
Breast_cancer_df["target"].value_counts()
```

```python
train_data = Breast_cancer_df.drop("target", axis=1)
train_data.head()
```

```python
target_data = Breast_cancer_df["target"]
target_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(train_data, target_data, test_size=0.2, random_state=20)
```

```python
print ('Number of training samples input is', X_train.shape)
print ('Number of testing samples input is', X_test.shape)
print ('Number of training samples output is', y_train.shape)
print ('Number of testing samples output is', y_test.shape)
```

```python
from sklearn.preprocessing  import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
#fitting the model
from sklearn.svm import SVC

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
prediction = svc_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report , accuracy_score
print(classification_report(y_test ,prediction))
```

```python
print(accuracy_score(y_test, prediction)*100)
```

```python
import pickle
with open('Breast_cancer_model.pkl', 'wb') as file:
    pickle.dump(svc_model, file)

with open('Breast_cancer_model.pkl', 'rb') as file:
    Breast_cancer_model = pickle.load(file)
```
